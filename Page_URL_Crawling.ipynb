```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import pandas as pd

# Chrome 드라이버 경로
driver_path = r'C:\Program Files (x86)\chromedriver_win32\chromedriver.exe'

# Selenium 웹 드라이버 설정
options = webdriver.ChromeOptions()
options.binary_location = r'C:\Program Files (x86)\Google\Chrome\Application\chrome.exe'
driver = webdriver.Chrome(executable_path=driver_path, options=options)

# 웹페이지 URL
url = 'http://www.jointips.or.kr/bbs/board.php?bo_table=team&sfl=wr_26&stx=2013'

# 페이지 로드
driver.get(url)

# 스크롤 다운
SCROLL_PAUSE_TIME = 1

# 스크롤 위치 초기화
last_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    # 페이지 스크롤 다운
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")

    # 페이지 로딩 대기
    time.sleep(SCROLL_PAUSE_TIME)

    # 스크롤 위치 확인
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# 페이지 HTML 가져오기
html = driver.page_source

# BeautifulSoup를 사용하여 HTML 파싱
soup = BeautifulSoup(html, 'html.parser')
```

```python
pd.set_option('display.max_colwidth', None)
# 데이터 추출
data = []
for element in soup.find_all('div', class_='name'):
    name = element.text.strip()
    url_element = element.find_next('div').find('a')
    if url_element is not None:
        url = url_element['href']
        data.append({'이름': name, 'URL': url})

# 데이터프레임 생성
df = pd.DataFrame(data)

# 테이블 출력
print(df)

# 드라이버 종료
driver.quit()
```

```python

# Excel 파일로 저장
output_file = 'URL_2013.xlsx'
df.to_excel(output_file, index=False)
```
