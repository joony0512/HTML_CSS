```python
# 웹페이지에 요청을 보내고 HTML 소스코드를 가져옵니다.
url = "http://www.jointips.or.kr/bbs/board.php?bo_table=team&wr_id=2170&sfl=wr_26&stx=2022"
response = requests.get(url)
html_content = response.text

# BeautifulSoup을 사용하여 HTML을 파싱합니다.
soup = BeautifulSoup(html_content, "html.parser")

# 원하는 정보를 추출하는 작업을 수행합니다.
# 예를 들어, <div> 태그의 클래스가 "content"인 요소들을 가져옵니다.
elements = soup.find_all("div", class_="content col-xs-12 col-sm-12 col-md-8")
df=[]
# 추출한 정보를 출력하거나 원하는 방식으로 가공합니다.
for element in elements:
  df.append(element.text)
    
data =df

# 데이터 추출
extracted_data = re.split(r'\n+|\t+', data[0])

# 데이터 테이블 생성
df = pd.DataFrame({'Data': extracted_data})



# 데이터 테이블 생성
df = pd.DataFrame({'Data': extracted_data})

# 필요한 인덱스 선택
indices = [1, 7, 9, 11, 18, 23]

# 특정 인덱스에 해당하는 행 추출
selected_rows = df.iloc[indices]

# 인덱스가 18인 경우 '\r' 제거
selected_rows.loc[18, 'Data'] = selected_rows.loc[18, 'Data'].replace('\r', '')

# 인덱스 재설정
selected_rows.index = ['사업장명', '사업분야1', '설립년월일', '대표자', 'tips선정일자', '운영기관']

# 설립년월일과 tips선정일자 데이터 처리
for col in ['설립년월일', 'tips선정일자']:
    date_str = selected_rows.loc[col, 'Data']
    date_str = date_str.replace('.', '')
    date_str = date_str.ljust(8, '0')  # 뒤에 0으로 채우기
    selected_rows.loc[col, 'Data'] = date_str

# 데이터프레임 출력
selected_rows
```

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

# List of URLs
url_list = URL_2013['URL']

# Empty lists to store extracted information
company_names = []
fields = []
establishment_dates = []
representatives = []
tips_selections = []
organizations = []

# Loop through each URL
for url in url_list:
    response = requests.get(url)
    html_content = response.text

    # BeautifulSoup을 사용하여 HTML을 파싱합니다.
    soup = BeautifulSoup(html_content, "html.parser")

    # 원하는 정보를 추출하는 작업을 수행합니다.
    elements = soup.find_all("div", class_="content col-xs-12 col-sm-12 col-md-8")

    # Empty list to store extracted data
    data = []

    # 추출한 정보를 출력하거나 원하는 방식으로 가공합니다.
    for element in elements:
        data.append(element.text)

    # 데이터 추출
    extracted_data = re.split(r'\n+|\t+', data[0])

    # 회사명 추출
    company_name = extracted_data[1]
    company_names.append(company_name)

    # 다른 정보 추출
    field = re.search(r'사업분야\n(.*?)\n', data[0])
    field = field.group(1).strip() if field else None
    fields.append(field)

    establishment_date = re.search(r'설립일\n(.*?)\n', data[0])
    establishment_date = establishment_date.group(1).strip() if establishment_date else None

    # 설립일을 'yyyymmdd' 형식으로 변환
    if establishment_date:
        parts = re.findall(r'\d+', establishment_date)
        if len(parts) >= 3:
            year = parts[0]
            month = parts[1].zfill(2)
            day = parts[2].zfill(2)
            establishment_date = f"{year}{month}{day}"
        else:
            establishment_date = "00000000"
    else:
        establishment_date = "00000000"
    establishment_dates.append(establishment_date)

    representative = re.search(r'대표자\n(.*?)\n', data[0])
    representative = representative.group(1).strip() if representative else None
    representatives.append(representative)

    tips_selection = re.search(r'TIPS 선정\n\r\n\t\t\t\t\t\t\t\t\t\t\t(.*?)\r\n', data[0])
    tips_selection = tips_selection.group(1).strip() if tips_selection else None
    

    def format_date(date_str):
      if len(date_str) >= 8:
        date_parts = date_str.split('.')
        year = date_parts[0]
        month = date_parts[1].zfill(2)
        day = date_parts[2].zfill(2) if len(date_parts) == 3 else '00'
        return f"{year}{month}{day}"
      else:
        year = date_str[:4]
        month = date_str[5:7] if len(date_str) >= 6 else '00'
        day = date_str[8:10] if len(date_str) >= 10 else '00'
        return f"{year}{month}{day}"


    # 'TIPS 선정' 컬럼 변환
    tips_selection = format_date(tips_selection)
    tips_selections.append(tips_selection)

    organization = re.search(r'운영기관\n([\s\S]*?)\n\n', data[0])
    organization = organization.group(1).strip() if organization else None
    organizations.append(organization)

# Create a dataframe with extracted information
df1 = pd.DataFrame({
    '회사명': company_names,
    '사업분야': fields,
    '설립일': establishment_dates,
    '대표자': representatives,
    'TIPS 선정': tips_selections,
    '운영기관': organizations
})

# Print the dataframe
print(df1)
```
